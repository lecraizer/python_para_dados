# Projeto - Biblioteca Virtual (POO)

Implemente as classes e métodos abaixo de forma objetiva.

1) **Classe `Livro`**  
   Atributos: `titulo`, `autor`, `ano_publicacao`, `disponivel=True`, `categoria` (string).  
   Método: `info()` → string com título, autor, ano e status disponível/não.

2) **Classe `Usuario`**  
   Atributos: `nome`, `email`, `livros_emprestados` (lista).  
   Método: `historico()` → se vazio, retorna mensagem simples; senão, lista de títulos.

3) **Classe `Biblioteca`**  
   Atributo: `catalogo` (lista de `Livro`).  
   Métodos:
   - `adicionar_livro(livro)`  
   - `listar_livros()` → imprime `info()` de cada livro; se vazio, mensagem simples.  
   - `emprestar_livro(titulo, usuario)` → empresta apenas se disponível; atualiza estados e retorna mensagem curta.  
   - `devolver_livro(titulo, usuario)` → devolve apenas se estava com o usuário; atualiza e retorna mensagem curta.  
   - `filtrar_por_categoria(categoria)` → retorna lista de livros cuja `categoria` contenha o texto (case-insensitive).

4) **Herança:** `LivroDigital(Livro)`  
   Atributo extra: `formato` (ex.: "PDF", "EPUB").  
   Sobrescreva `info()` para incluir `formato` em maiúsculas.

5) **Cenário rápido (integração)**  
   Crie uma `Biblioteca`, alguns `Livro`/`LivroDigital` com categorias distintas; adicione ao catálogo; realize um empréstimo e uma devolução; liste livros antes/depois; mostre um filtro por categoria.

6) **Devolução de Livros:**
Adicione na classe `Biblioteca` o método `devolver_livro(titulo, usuario)` que:
- Só devolve se o livro estiver emprestado **e** o título constar em `usuario.livros_emprestados`;
- Ao devolver: marca o livro como disponível e remove o título do histórico do usuário;
- Retorna uma mensagem curta de confirmação ou `"Erro na devolução."`.

7) **Adicionando Categorias de Livros:**
Altere a classe `Livro` para incluir o atributo `categoria` (string).  
Na classe `Biblioteca`, implemente `filtrar_por_categoria(categoria)` que:
- Faz busca **case-insensitive** em `livro.categoria`;
- **Retorna** a lista de objetos `Livro` cujo campo `categoria` contém o texto informado.
